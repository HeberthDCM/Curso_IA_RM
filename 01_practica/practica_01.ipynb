```python
import pandas as pd
import seaborn as sb
import statsmodels.api as sm


datos_persona = pd.read_csv('alturas-pesos.csv')
sb.scatterplot(x='Peso', y='Altura', data = datos_persona, palette ='deep', hue='Sexo')
```

Modelo

```python
datos_persona
datos_persona.head()
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
peso = datos_persona['Peso']
peso_procesado = peso.values.reshape(-1,1) 
altura_procesada = datos_persona['Altura'].values.reshape(-1,1)
# entrenamos el modelo
modelo.fit(altura_procesada,peso_procesado)
```

```python
# probamos el modelo
modelo.predict([[182]])
```

```python
modelo.score(altura_procesada,peso_procesado)
```

```python
#estadisticos
X_constante =sm.add_constant(altura_procesada)
modelo_stat = sm.OLS(peso_procesado,X_constante).fit()
print(modelo_stat.summary())
```
